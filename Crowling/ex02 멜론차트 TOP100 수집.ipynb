```python
soup = bs(res.text,"lxml")
```

# 노래제목 수집

```python
# 노래제목에 해당하는 태그정보를알아온 다음에.
# soup.select() 노래제목만 수집
# 클래스 값 사이에 공백이 있다는 의미는 클래스가 복수개
# 공백을 지우고 .을 통해서 클래스를 이어준다.
title = soup.select("div.ellipsis.rank01>span>a")
```

```python
for i in title:
    print(i.text)
```

```python
# 수집한 데이터를  사용자가 사용할 수 있게 변형(태그를 삭제, 글자만 저장)
# 순수한 텍스트 정보만 담긴 리스트 새로 제작
singerList = [] #순수한 가수이름정보
titleList = [] #순수한 노래제목정보

for i in range(len(singer4)):
    singerList.append(singer4[i].text)

for i in range(len(title)):
    titleList.append(title[i].text)
```

```python
# 변형된 데이터를 표형식으로 제작(시각화)
import pandas as pd
```

```python
dic = {"가수":singerList, "노래제목":titleList}
```

```python
melon = pd.DataFrame(dic)
```

```python
# 표를 파일로 저장
# 나 melon이라는 표를 csv 형태로 저장 해줘~
# 한글을 저장할 때는 인코딩이 필수!
melon.to_csv("멜론차트.csv", encoding="euc-kr")
```
