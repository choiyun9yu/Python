```python
# 소수 판독기
def isPrime(x):
    for i in range(2, x):
        if x%i==0:
            return False
        return True
        
a = [12, 13, 7, 9, 19]

for j in a:
    r=isPrime(j)
    print(r, end=' ')
```

```python
# 사용자 함수
def plus_one(x):
    return x+1
print(plus_one(1))

# 람다 함수 (익명의 함수 or 람다 표현식이라고도 불림)
# 람다함수는 호출하려면 변수에 할당을 해야한다
pluse_two = lambda x: x+2
print(pluse_two(1))

# 람다함수는 내장함수의 인자로 사용될 때 편리하다
a=[1,2,3]
print(list(map(plus_one,a)))

```

```python
# 정사면체, 정육면체 주사위 합의 경우의 수 카운팅
n, m = 4, 6
# 경우의 수 총 24가지
# 근데 중복 제거하려면?
# 그냥 생각하면 제일 작은데 1+1이라서 2고 제일큰게 4+6이라서 10 9개
cnt = [0]*(n*m)
max = -float('inf')
for i in range(1, n+1):
    for j in range(1, m+1):
        cnt[i+j] += 1
for i in range(n+m+1):
    if cnt[i]>max:
        max=cnt[i]
for i in range(n+m+1):
    if cnt[i] == max:
        print(i, end=' ')
```
