```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 1. req를 통해서 리뷰사이트 요청
# 주의점! 응답코드가 200인지 꼭 확인하기!
res = req.get("https://movie.naver.com/movie/bi/mi/point.naver?code=81888")
```

```python
# 2. 컴퓨터가 이해할 수 있는 언어로 변경(bs)
soup = bs(res.text, "lxml")
```

```python
# 3. soup데이터에서 select를 통해 리뷰 정보만 수집
reviews = soup.select("div.score_reple > p > span") 
```

### iframe의 개념!
- iframe은 창을 통해서 다른 서버에 있는 정보를 보여줄 때 사용
- 찾는 방법은 
- 1. f12개발자 도구에서 (ctrl + F) iframe을 검색
- 2. 실제 사용되는 iframe 찾기 (화면에 색이 파란색으로 변경)
- 3. 실제 주소로 찾아가기! (src = "url값")
- 3.1 url이 폴더의 경로만 있는 경우에는 http://서버주소 까지 입력해줘야함

```python
res1 = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false")
```

```python
soup1= bs(res1.text, "lxml")
```

```python
reviews1 = soup1.select("div.score_reple > p > span")   # 관람객 + 리뷰 데이터
```

```python
# 불필요한 개행을 지우는 방법 .strip()
for i in reviews1:
    print(i.text.strip())   
```

```python
# 불필요한 관람객이라는 글자들만 삭제(추출)
# extract() -> soup 데이터에 해당하는 요소들을 추출(제거)한다. // str에만 extract 가능한 듯
ico = soup1.select("span.ico_viewer")
for i in ico:
    i.extract()   # 추출 함수 .extract()
```

# 1page부터 10page 까지 수집

```python
for i in tq(range(1,100)):
    res = req.get(f"https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={i}")
    soup = bs(res.text,"lxml")
    ico = soup.select("span.ico_viewer")
    for i in ico :
        i.extract() 
    review = soup.select("div.score_reple > p > span")
    for i in review :
        print(i.text.strip())
```
