## 동기 프로그래밍

```python
import time

def find_users_sync(n):
    for i in range(1, n+1):
        print(f'{n}명 중 {i}번째 사용자 검색 중...')
        time.sleep(1)
    
    print(f'>> 총 {n}명의 사용자 조회 완료')

find_users_sync(5)
```

- 3개의 요청을 동기 처리

```python
def process_sync() :
    start = time.time()
    find_users_sync(3)
    find_users_sync(2)
    find_users_sync(1)
    end = time.time()
    print(f'>> 동기처리 방식의 시간 : {end-start}')
    
process_sync()
```

## 비동기 프로그래밍

```python
#비동기에서는 동시에 여러개의 작업을 하는데 작업들 간에 메세지를 교환하는 이벤트 루프
!pip install nest_asyncio
```

```python
import asyncio #비동기용 입출력 라이브러리
import time

# async : 비동기 함수 선언 (코루틴-coroutine 함수)
# await : 비동기 함수로 된 함수를 호출
async def find_users_sync(n):
    for i in range(1, n+1):
        print(f'{n}명 중 {i}번째 사용자 검색 중...')
        await asyncio.sleep(1)
    
    print(f'>> 총 {n}명의 사용자 조회 완료')

find_users_sync(5)  #비동기라서 불러오면 코루틴 객체임
```

```python
import nest_asyncio 
import asyncio

# 이벤트 루프(비동기 함수 처리할 때 가장 먼저 그리고 항상 해줘야하는 것)
nest_asyncio.apply()

async def process_async() :
    start = time.time()
    #await() : async로 선언된 비동기 함수를 호출하는 함수
    #여러개 함수 호출 할 땐 'awiat asyncio.wait()'를 사용해서 비동기 함수 호출
    await asyncio.wait([find_users_sync(3), find_users_sync(2), find_users_sync(1)])
    end = time.time()
    print(f'>> 동기처리 방식의 시간 : {end-start}')
    
# 파이썬 3.7부터 비동기 함수를 실행해주는 run() 함수
asyncio.run(process_async())
    
# loop = asyncio.get_event_loop()
# loop.run_until_complete(process_async())
```

  - boundary : 복합문서 내의 각 문서들을 구분하는 분리자 (동영상이므로 frame으로 구분)

- 이미지 전송 부분을 동기방식으로 구현하는 경우

```python
from flask import Flask, render_template, Response
import cv2

cap = cv2.VideoCapture(0)
# 이미지를 읽어서 한프레임씩 바로 전송하는 스트리밍 함수
def get_frames() :
    while True :
        ret, frame = cap.read()
        # 현재 프레임이미지를 읽지 못했다면 다음 프레임이미지를 읽는다

        if not ret :
            continue
        else :
            ret1, buffer = cv2.imencode(".jpg", frame)
            # 전송하기 위해 바이트로 변환
            frame = buffer.tobytes()
            # 전송
            yield (b'--frame\r\n' b'Content-Type: image/jpeg\r\n\r\n' + frame + b'\r\n')

app = Flask(__name__)

# 초기화면
@app.route("/")
def index() :
    return render_template("index.html")

@app.route("/video_feed")
def video_feed() :
    return Response(get_frames(), mimetype='multipart/x-mixed-replace; boundary=frame')

if __name__ == "__main__" :
    app.run(host="127.0.0.1", port=9000)
```

- 이미지 전송 부분을 비동기 방식으로 구현한 경우

```python
import cv2
import nest_asyncio
import uvicorn
from fastapi import FastAPI, Request
from fastapi.templating import Jinja2Templates
from fastapi.responses import StreamingResponse  #fastAPI에서는 리스폰스 형태마다 다 다르게 지정해야함 / 요청받는 것도 다 다르게
import asyncio

# 이 두개는 비동기할 때 고정
app = FastAPI()
nest_asyncio.apply()

# directory : jinja템플릿이 퐇마된 html문서의 위치 지정 해줘야함
templates = Jinja2Templates(directory="templates")

cap = cv2.VideoCapture(0)

#이미지 읽어서 한 프레임씩 바로 전송하는 스트리밍 함수 만들기

def get_frames(): 
    while True:
        ret, frame = cap.read()

        #스트리밍 서비스 이기 때문에 프레임 이미지를 읽지 못했다면 다음 프레임 이미지를 읽도록 하기 위해 컨티뉴~
        if not ret :
            continue
        else:
            ret,buffer = cv2.imencode('.jpg', frame)
            #전송하기 위해서 바이트로 변환
            frame = buffer.tobytes()

            #전송
            yield (b'--frame\r\n' b'Content-Type: image/jpeg\r\n\r\n' + frame + b'\r\n')
            
@app.get('/')
async def index(request : Request): #어디서 리퀘스트 받는지 받아둠
    # 보낼 request를 설정
    return templates.TemplateResponse("index.html", {'request': request})

@app.get("/video_feed")
async def video_feed():
    return StreamingResponse(get_frames(), media_type='multipart/x-mixed-replace; boundary=frame')
            
if __name__ == "__main__":
    # soket.send() raised exception. 어떻게 해결?
#     await asyncio.sleep(1) 
    uvicorn.run(app, host='127.0.0.1', port=9000, debug=False)
```

```python
import cv2
import nest_asyncio
import uvicorn
from fastapi import FastAPI, Request
from fastapi.templating import Jinja2Templates
from fastapi.responses import StreamingResponse  #fastAPI에서는 리스폰스 형태마다 다 다르게 지정해야함 / 요청받는 것도 다 다르게
import asyncio

block_size = 21
# 결과값에서 빼는 수치
C = 8

# 이 두개는 비동기할 때 고정
app = FastAPI()
nest_asyncio.apply()

# directory : jinja템플릿이 퐇마된 html문서의 위치 지정 해줘야함
templates = Jinja2Templates(directory="templates")

cap = cv2.VideoCapture(0)

#이미지 읽어서 한 프레임씩 바로 전송하는 스트리밍 함수 만들기

def get_frames(): 
    while True:
        ret, frame = cap.read()
        frame=cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        frame = cv2.adaptiveThreshold(frame, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                               cv2.THRESH_BINARY, block_size, C)

        #스트리밍 서비스 이기 때문에 프레임 이미지를 읽지 못했다면 다음 프레임 이미지를 읽도록 하기 위해 컨티뉴~
        if not ret :
            continue
        else:
            ret,buffer = cv2.imencode('.jpg', frame)
            #전송하기 위해서 바이트로 변환
            frame = buffer.tobytes()

            #전송
            yield (b'--frame\r\n' b'Content-Type: image/jpeg\r\n\r\n' + frame + b'\r\n')
            
@app.get('/')
async def index(request : Request): #어디서 리퀘스트 받는지 받아둠
    # 보낼 request를 설정
    return templates.TemplateResponse("index.html", {'request': request})

@app.get("/video_feed")
async def video_feed():
    return StreamingResponse(get_frames(), media_type='multipart/x-mixed-replace; boundary=frame')
            
if __name__ == "__main__":
    # soket.send() raised exception. 어떻게 해결?
#     await asyncio.sleep(1) 
    uvicorn.run(app, host='127.0.0.1', port=9000, debug=False)
```
