## 색상공간 (Color Space) 이해하기

- RGB : Red, Green, Blue로 구성된 색상공간 - PC의 모니터에 주로 사용
  - 가산혼합 : 색을 섞을수록 명도가 높아지는 혼합  
<img src="./data/lecture_image/07_color_space2.png" width=40%>


- CMYK : Cyan, Magenta, Yellow, Black으로 구성된 색상공간 - 인쇄용으로 주로 사용
  - 감산혼합 : 색을 섞을수록 명도가 낮아지는 혼합  
<img src="./data/lecture_image/07_color_space3.png" width=40%>  


- HSV : Hue(색상), Saturation(채도), Value(명도)로 변환한 색상공간 
  - 직관적으로 색상 파악이 용이  
<img src="./data/lecture_image/07_color_space.png" width=50%>

- YUV : Y (밝기, 휘도), U, V (색상)을 이용해서 색상을 표현하는 색상공간
  - U축 : 파란색에서 밝기 성분을 뺀 값 (U = B - Y)
  - V축 : 빨간색에서 밝기 성분을 뺀 값 (V = R - Y)
  - 아날로그 컬러신호 변환에 주로 사용
  - TV나 비디오카메라에 주로 사용
  
  
- YCbCr : Y (밝기, 휘도), Cb (청색 크로마 성분), Cr (적색 크로마 성분)을 이용해서 색상을 표현하는 색상공간
   - YUV와 유사한 형태로 RGB 색공간과는 다르게 별도의 명암을 구분하는 U(Cb), V(Cr) 사용
   - 디지털 텔레비전, 비디오, 디지털 사진에서 사용하는 색공간  
   <img src="./data/lecture_image/07_color_space4.png" width=20%>   


- CIE : 인간의 색채 인지에 대한 연구와 측정을 바탕으로 수학적으로 정의된 색상 공간
  - XYZ : 적색(X), 녹색(Y), 청색(Z) 계열의 반응정도값을 적용 
  - Lab : 색오차와 근소한 색차이를 표현하기 위해 변환된 색상 공간
  - Luv : 지각적 분균등성을 해소한 균등 색상 공간 (3차원 직교좌표를 이용하는 색상 공간)  
  <img src="./data/lecture_image/07_color_space5.png" width=30%>

## 색상 추적


```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("../OpenCV_data/image/balloon.jpg", cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.xticks([]); plt.yticks([])
plt.imshow(img)
plt.show()
```

```python
import numpy as np

# 초록 풍선 찾기
# RGB를 HSV로 변환
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

# HSV 공간에서 빨강색 영역을 설정
# 색상, 채도, 명도 순
lower = np.array([40, 70, 100])
higher = np.array([80, 255, 255])

# HSV 색상 공간에 범위 값을 마스크로 생성
mask = cv2.inRange(hsv, lower, higher)

#풍선 찾기
img2 = cv2.bitwise_and(img, img, mask=mask) #(1, 2, 3)  -> 2와 3을 and 연산하고 1에 넣어줘라
plt.xticks([]); plt.yticks([])
plt.imshow(img2)
plt.show()
```

```python
img = cv2.imread("../OpenCV_data/image/candies.png", cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.xticks([]); plt.yticks([])
plt.imshow(img)

# RGB를 HSV로 변환
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

# HSV 공간에서 빨강색 영역을 설정
# 색상, 채도, 명도 순
lower = np.array([47, 70, 50])
higher = np.array([90, 255, 255])

# HSV 색상 공간에 범위 값을 마스크로 생성
mask = cv2.inRange(hsv, lower, higher)

img2 = cv2.bitwise_and(img, img, mask=mask) #(1, 2, 3)  -> 2와 3을 and 연산하고 1에 넣어줘라
plt.xticks([]); plt.yticks([])
plt.imshow(img2)
plt.show()
```

## 이미지 연산


### 이미지 필터링 - blur

- <font color=red>cv2.filter2D(img, -1, kernel)</font>
  - 이미지와 필터를 컨볼루션 연산 수행
  - 2번째 파라미터는 색의 깊이 (몇 비트)로 -1이면 원본 이미지와 동일
  - kernel : 적용할 2차원 필터
  
 <img src="./data/lecture_image/07_filter2d.png" width=60%></td></tr>    
 
 - Convolution 연산 : 같은 위치끼리 곱한다 -> 곱한것들을 다 더한다  / -12
 - 어떤 필터를 적용하느냐에 따라 이미지가 가진 특성을 뽑아낼 수 있다.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../OpenCV_data/image/suzy2.jpeg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#필터설정
kernel = np.ones((5,5), np.float32)/ 25

kernel = np.array([[1,0,0],
                 [0,1,0],
                 [0,0,1]])
print(kernel)
```

### 컨볼루션 연산

```python
img2 = cv2.filter2D(img, -1, kernel)

plt.subplot(1,2,1)
plt.xticks([]); plt.yticks([])
plt.title('nomal_SUZY')
plt.imshow(img)

plt.subplot(1,2,2)
plt.xticks([]); plt.yticks([])
plt.title('light_SUZY')
plt.imshow(img2)
plt.show()
```

### 모폴로지 - 침식과 팽창 

- Morphology : 영상의 형태학적 측면을 다루는 기법으로 전처리 등에 사용
   - 


- Erosion (침식) : 이미지의 경계부분을 배경 픽셀로 변경하는 작업 (1이 0으로 변경)
- Dilation (팽창) : 이미지의 배경부분을 전경 픽셀로 변경하는 작업 (0이 1로 변경)
  - iterations = 1 : 반복 적용 회수 (회수가 많아지면 침식과 팽창의 효과가 커짐)  
 <img src="./data/lecture_image/07_mophology.png" width=60%></td></tr>    

```python
img = cv2.imread('../OpenCV_data/image/number.png', cv2.IMREAD_GRAYSCALE)

plt.xticks([]); plt.yticks([])
plt.imshow(img, cmap='gray')
plt.show()
```

#### 침식

```python
#침식
kernel = np.ones((3,3), np.uint8)

# interations : 수행 횟수 (클수록 글씨들이 얇아짐)
result_e = cv2.erode(img, kernel, iterations=1)

plt.xticks([]); plt.yticks([])
plt.imshow(result_e, cmap='gray')
plt.show()
```

#### 팽창

```python
#팽창 (마스크 크게 만들 수 있음, 마스크 노이즈 제거 가능)
kernel = np.ones((3,3), np.uint8)

# deilate : 수행 횟수 (클수록 글씨들이 두꺼워짐)
result_d = cv2.dilate(img, kernel, iterations=1)

plt.xticks([]); plt.yticks([])
plt.imshow(result_d, cmap='gray')
plt.show()
#글씨 마스크 만들 수 있음
```

### Opening과 Closing

- Opening : erosion 수행 후에 바로 dilate을 수행하여 원래 이미지 크기로 돌려 놓는 것
    - 배경(검정)에 노이즈(흰색)가 있는 경우 배경에 있는 흰색을 제거하는 역할  
<img src="./data/lecture_image/07_mophology3.png" width=40%>


- Closing : dilate 수행 후에 바로 erosion을 수행하여 원래 이미지 크기로 돌려 놓는 것
    - 글쓰(흰색)에 노이즈(검정)가 있는 경우 글씨에 있는 검정색을 제거하는 역할  
<img src="./data/lecture_image/07_mophology4.png" width=40%>  

#### 오프닝 (글씨 외부 노이즈 제거 | 침식 -> 팽창)

```python
#Opening
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_a = cv2.imread('../OpenCV_data/image/noise_A.png', cv2.IMREAD_GRAYSCALE)

plt.subplot(1,2,1)
plt.xticks([]); plt.yticks([])
plt.imshow(img_a, cmap='gray')
plt.show()
```

```python
img_a = cv2.imread('../OpenCV_data/image/noise_A.png', cv2.IMREAD_GRAYSCALE)

plt.subplot(1,2,1)
plt.xticks([]); plt.yticks([])
plt.imshow(img_a, cmap='gray')

#침식
kernel = np.ones((3,3), np.uint8)
# interations : 수행 횟수 (클수록 글씨들이 얇아짐)
img3 = cv2.erode(img_a, kernel, iterations=1)

#팽창 (마스크 크게 만들 수 있음, 마스크 노이즈 제거 가능)
kernel = np.ones((3,3), np.uint8)
# deilate : 수행 횟수 (클수록 글씨들이 두꺼워짐)
img3 = cv2.dilate(img3, kernel, iterations=1)

plt.subplot(1,2,2)
img3 = cv2.erode(img3, kernel, iterations=1)
plt.xticks([]); plt.yticks([])
plt.imshow(img3, cmap='gray')
plt.show()
```

#### 클로징 (글씨 내부 노이즈 제거 | 팽창 -> 침식)

```python
img_b = cv2.imread('../OpenCV_data/image/noise_B.png', cv2.IMREAD_GRAYSCALE)

plt.subplot(1,2,1)
plt.xticks([]); plt.yticks([])
plt.imshow(img_b, cmap='gray')

#팽창 (마스크 크게 만들 수 있음, 마스크 노이즈 제거 가능)
kernel = np.ones((3,3), np.uint8)
# deilate : 수행 횟수 (클수록 글씨들이 두꺼워짐)
img4 = cv2.dilate(img_b, kernel, iterations=2)

#침식
kernel = np.ones((3,3), np.uint8)
# interations : 수행 횟수 (클수록 글씨들이 얇아짐)
img4 = cv2.erode(img4, kernel, iterations=2)

plt.subplot(1,2,2)
img4 = cv2.erode(img4, kernel, iterations=1)
plt.xticks([]); plt.yticks([])
plt.imshow(img4, cmap='gray')
plt.show()
```
