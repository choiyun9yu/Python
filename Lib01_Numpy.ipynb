### 1차원 배열

```python
list1= [1,2,3,4,5]
print(type(list1))

# list to array
arr1 = np.array(list1)
print(type(arr1))

# array는 자료형이 같아야해서 1,2,3이 str으로 변한다
arr0=np.array([1,2,3, "야호!"])
print(type(arr0[1]))

# array to list
print(type(arr0.tolist()))

print('-'*115)

# 1차원 배열 생성 (n) : 0~n-1
arr_ar=np.arange(10)
print(arr_ar)

print('-'*115)

# 1차원 배열 생성 후 reshape 생성 갯수랑 행x열 이 같아야한다.
arr_ar2=np.arange(10).reshape(5,2)
print(arr_ar2)
```

### 2차원 배열

```python
arr2 = np.array([[1,2,3,], [4, 5, 6]])

# 배열의 형태 확인(행, 열)
print("배열의 형태 :", arr2.shape)

# 배열의 요소 전체 개수
print("배열의 전체 요소개수 :", arr2.size)

# 배열의 데이터 타입
print("배열의 데이터 타입 :", arr2.dtype)   #int64의 의미는 2**64 = -2**32 부터 +2**32-1까지 
                    #32는 bit(0,1) 2의 32제곱 의미
                    #int8 -> 1비트당 2개의 숫자를 표현할 수 있음 2**8 = 256개 표현 가능 (-128~127)
                    #int32 -> -2,147,483,648 ~ 2,147,483,647
                    
# 배열의 차원 확인
print("배열의 차원 :", arr2.ndim)
```

```python
# 0으로 배열 생성(차원, 행, 열) // 두개만 적으면 차원 빼고 (행,열)로 인식한다.
arr_zeros=np.zeros((3,4))
print(arr_zeros)

print('-'*115)

# 1로 배열 생성(차원, 행, 열)
arr_ones=np.ones((2,4,3))
print(arr_ones)

print('-'*115)

# 특정값으로 배열 생성하기 ((차원, 행, 열), 특정값)
arr_full = np.full((3,4,3),2)
print(arr_full)

print('-'*115)

# 범위로 배열 생성 (시작값, 미만값, 증감폭)
arr_range=np.arange(1,10)
print(arr_range)
```

```python
# 배열 행렬 전치 
arr3=np.zeros((2,4))
print(arr3)
arr_t=arr3.T  # 전치를 위해서는 대입 연산자 사용해야함
print(arr_t)

print('-'*115)

# 배열의 행과 열 재배치
arr3.reshape(2,4)
print(arr3)

print('-'*115)

# 자료형태 변경하기 np.array(listName, dtpye=np.자료형)
arr_ch=np.array(arr3, dtype=np.str)
print(type(arr_ch[0][0]))
```

```python
# 랜덤 난수(0~1) 생성하기 (행,열)
arr_rand=np.random.rand(2,3)
print(arr_rand)

print('-'*115)

# 랜덤 정수 1개 생성하기 (미만값)
arr_randint0=np.random.randint(10)
print(arr_randint0)

print('-'*115)

# 1차원 랜덤 정수 배열 생성하기 (미만값, size=개수)
arr_randint1=np.random.randint(10, size=4)
print(arr_randint1)

print('-'*115)

# 2차원 랜덤 정수 배열 생성하기 (미만값, size=(행,열))
arr_randint2=np.random.randint(10, size=(3,2))
print(arr_randint2)
```

### 배열의 연산

```python
list4 = [[1,2],
          [3,4]]
print(list4+list4)
```

```python
# 배열 사칙연산 (덧셈 곱셈 뺄셈 나눗셈 다 됨)
arr4 = np.array([[1,2],
                 [3,4]])
print(arr4+arr4)

# 합계 (배열 요소 다 합침)
print('합계', np.sum(arr4))
print('합계', arr4.sum())

# 평균
print('평균', np.mean(arr4))
print('평균', arr4.mean())

# 중앙값
print('중앙값', np.median(arr4))

# 제곱근
print('제곱근') 
print(np.sqrt(arr4))

# 절대값
print('절대값') 
print(np.abs(arr4))
```

### 인덱싱/슬라이싱

- 인덱싱 : 배열 중 하나의 인덱스값에 있는 데이터를 가져오는 것
- 슬라이싱 : 배열의 범위를 지정하여 여러개의 데이터를 가져오는 것

```python
# 인덱싱
arr5 = np.array([[1,2,3],
                 [4,5,6]])
print(arr5[0][1])

print('-'*115)

# 슬라이싱
arr6 = np.arange(10)
print(arr6[3:8])

print('-'*115)

# 2차원 데이터 슬라이싱(행시작:행끝, 열시작,열끝)
arr7=np.arange(50).reshape(5,10)
print(arr7)

print('-'*115)

print(arr7[0:2, 0:2])
```

```python
arr19 = np.arange(18).reshape(3,6)
arr19
```

### 불리언 인덱싱

- 특정 조건에 맞으면 True, 아니면 False 값으로 배열을 만들어 조건에 맞는 값에 접근하는 방법

```python
arr10 = np.array(np.random.randint(50,100, size=8))
print(arr10)

#boolean(불리언) : True(참) 또는 False(거짓)으로 출력되는 자료형태
print(arr10 > 60)

# 불리언 인덱싱
arr10[arr10 > 60]
```

## 예제) BMI지수 및 과체중 BMI값 확인

#### 데이터 로드

```python
#delimiter(구분자) : 파일에서 데이터를 구분해주는 문자 지정
data = np.loadtxt('rowdata/height_weight.txt', delimiter=",")
print(data)  # 키는 cm / 몸무게는 kg
```

#### 데이터의 속성확인
- 배열 형태
- 배열 차원수
- 배열 전체요소 개수

```python
print("배열 형태:",data.shape)
print("배열 차원수:",data.ndim)
print("배열 전체요소 개수:",data.size)
```

#### 데이터에서 키와 몸무게 데이터들을 각각 변수에 담기

```python
#키
height = data[0]
#몸무게
weight = data[1]

print(height)
print(weight)
```

```python
bmi=weight/((height/100)**2)
print(bmi)
```

#### BMI지수가 23이상 25이하인 사람은 몇 명일까요?
- 불리언 인덱싱에서는 and -> & , or -> |, not->~를 쓴다

```python
bmi[(bmi >= 23) & (bmi <= 25)]
```

```python
len(bmi[(bmi >= 23) & (bmi <= 25)])
```

```python
user_mean_list = []

for i in user_id:
    # 1번 ~6040번 사용자가 매긴 평점들의 평균을 각각 구해주는 코드
    user_mean = movie[ movie[:,0]==i][:,2].mean()
    # 빈 리스트에 사용자 번호와 그 사람이 매긴 평점의 평균을 리스트로 저장
    user_mean_list.append([i,user_mean])
user_mean_list
```
